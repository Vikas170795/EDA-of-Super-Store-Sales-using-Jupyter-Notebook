# Visualization of data

# Bar plot between category vs sub-category

```python
plt.figure(figsize =(16,10))
plt.ylabel("Category",color ="blue")
plt.xlabel("Sub-Category",color ="blue")
plt.bar("Sub-Category","Category",data = df_data)
plt.show()
```

```python
#In furniture category,the frequency of individual category is very less
#In Office supplies,the frequency of individual category is Medium
#In Technology category,the frequency of individual category is High
```

# Pie plot of Sub_category

```python
plt.figure(figsize=(12,10))
df_data["Sub-Category"].value_counts().plot.pie(autopct ="%1.1f%%")
plt.show()
```

# Count plot of categories

```python
print(df_data["Category"].value_counts())
plt.figure(figsize=(10,7))
plt.title("Count plot of Categories",color='blue')
sn.countplot(x =df_data["Category"])
plt.xticks(rotation= 90)
plt.show()
```

# Count plot of Sub-categories

```python
print(df_data["Sub-Category"].value_counts())
plt.figure(figsize=(15,8))
plt.title("Count plot of sub-categories",color ='orange')
sn.countplot(x =df_data["Sub-Category"])
plt.xticks(rotation= 90)
plt.show()
```

# Plot of total sales per region

```python
print(df_data.groupby('Region')['Sales'].agg(['sum']))
df_data.groupby('Region')['Sales'].agg(['sum']).plot.bar()
plt.title("Total sales per region",color='orange')
plt.figure(figsize=(10,7))
plt.show()
```

# Plot of total sales for each type of shipping mode

```python
print(df_data.groupby('Ship Mode')['Sales'].agg(['sum']))
df_data.groupby('Ship Mode')['Sales'].agg(['sum']).plot.bar()
plt.title("Total sales per Ship Mode",color='orange')
plt.figure(figsize=(10,7))
plt.show()
```

# Plot of sales for each segment type

```python
print(df_data.groupby('Segment')['Sales'].agg(['sum']))
df_data.groupby('Segment')['Sales'].agg(['sum']).plot.pie(subplots="True",autopct="%1.1f%%")
plt.title("Total sales per Segment",color='orange')
plt.figure(figsize=(10,7))
plt.show()
```

# Heatmap for corelation among the columns

```python
plt.figure(figsize=(10,7))
sn.heatmap(df_data.corr(),annot=True)
plt.show()
```

# Heatmap for covariance among the columns

```python
plt.figure(figsize=(10,7))
sn.heatmap(df_data.cov(),annot=True)
plt.show()
```
